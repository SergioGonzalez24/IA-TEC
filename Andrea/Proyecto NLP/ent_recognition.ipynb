```python
import spacy
from spacy import displacy
import pandas as pd
```

```python

nlp = spacy.load("es_core_news_sm")
```

```python

oracion = "La científica Marie Curie nació en Varsovia, Polonia, el 7 de noviembre de 1867, y fue galardonada con el Premio Nobel de Física en 1903 y el Premio Nobel de Química en 1911 por sus contribuciones pioneras en los campos de la radiactividad y la fisicoquímica."

# Procesar el texto con spaCy
doc = nlp(oracion)
```

```python

df_tokens = []
for token in doc:
    df_tokens.append({"Texto": token.text, "Parte de la oración": token.pos_, "Dependencia": token.dep_})
    

df = pd.DataFrame(df_tokens)


print(df)
```

```python

# Imprimir entidades nombradas en una tabla
entidades = []
for entidad in doc.ents:
    entidades.append((entidad.text, entidad.label_))

df = pd.DataFrame(entidades, columns=["Entidad", "Tipo"])
print("Tabla de Entidades:")
print(df)
```

```python
displacy.render(doc,style='ent', jupyter=True, options={'distance': 90})
```

```python
displacy.render(doc,style='dep' ,jupyter=True, options={'distance': 90})
```
