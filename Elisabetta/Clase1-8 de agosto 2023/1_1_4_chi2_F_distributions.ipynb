## $\chi^2$ distribution
https://en.wikipedia.org/wiki/Chi-squared_distribution

The probability density function of the distribution is:

$f(x,k)=\displaystyle\frac{x^{k/2-1}e^{-x/2}}{2^{k/2}\Gamma(k/2)}\,\,\,\,\,\,x>0$

$f(x,k)=0\,\,\,\,otherwise$


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/3/35/Chi-square_pdf.svg/482px-Chi-square_pdf.svg.png" style="margin:10px auto 20px; display:block; width:720px">

In statistics, this distribution is important because the $\chi^2$ distribution with $k$ degrees of freedom is the distribution of the sum of the squares of $k$ indipendent standard normal random variables.


If $Z_1,Z_2,...,Z_k$ are independent, standard normal random variables and Q is the sum of their squares:

$Q=\displaystyle\sum_{i=1}^k Z_i^2$

then $Q\sim\chi^2_k$

```python
from scipy.stats import chi2
```

```python
chi2.cdf(12, df=5)
```

## F distribution

https://en.wikipedia.org/wiki/F-distribution

The F distribution depends on two parameters (two degrees of freedom) and it is important in statistics in ANOVA tests. 
The distribution describes the random variable:

$X=\frac{S_1/d_1}{S_2/d_2}$

where $S_1,S_2$ are two random variables with $\chi^2$ distribution with $d_1,d_2$ degrees of freedom respectively.


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/74/F-distribution_pdf.svg/650px-F-distribution_pdf.svg.png" style="margin:10px auto 20px; display:block; width:720px">

```python
from scipy.stats import f
```

```python
f.cdf(3.5, dfn=4, dfd=26)
```
