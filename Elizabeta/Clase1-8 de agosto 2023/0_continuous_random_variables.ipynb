## 1. Variable aleatoria

> Es una descripción numérica del resultado de un experimento.

**Observaciones.**

- Más formalmente, una variable aleatoria es una función $X : \Omega \longrightarrow \mathbb{R}$. <br/> (*Las variables aleatorias deben tomar valores numéricos.*)
- **El rango de $X$ es el conjunto de todos los posibles valores que toma $X$.**

### Variables aleatorias continuas

![](https://drive.google.com/uc?export=view&id=1f3x8a2Rv5Vqb7WmRe4A7I0YneOyCe_LE)

$$\text{El rango de } X \text{ no se puede enumerar.}$$

**Ejemplo.**

En el experimento aleatorio de medir la vida útil (en horas) de un dispositivo electrónico, donde

$$\Omega = [0, \infty)$$

definimos la variable aleatoria $X$ como la **duración del dispositivo** (en horas), y la representamos como sigue

$$X(w) = w.$$

El rango de $X$ es $[0,\infty)$. Entonces $X$ es una variable continua.

## 2. Variable aleatoria continua

### 2.1. Función de densidad

Una función $f(x)$ es la función de densidad de la variable continua $X$ si satisface
1. $f(x) \geq 0$ para todo $x$ en el rango de $X$
2. $\displaystyle \int_{-\infty}^{\infty} f(x)\,\mathrm{d}x = 1$ (*el área bajo la gráfica de $f$ es 1*)
3. $\displaystyle P[a \leq X \leq b] = \int_{a}^{b} f(x)\,\mathrm{d}x$, para cualquier intervalo $[a,b]$ contenido en el rango de $X$

**Observación.**
> *La propiedad 3 en la definición anterior nos dice que la probabilidad que $a \leq X \leq b$ es igual al área bajo la gráfica de la función de densidad de $X$ comprendida entre $x=a$ y $x=b$.*

**Observación.**

- Si $X$ es una variable continua, entonces $P[X=c] = 0$
- Si $X$ es una variable continua, entonces

$$P[a\leq X \leq b] = P[a \leq X < b] = P[a < X \leq b] = P[a < X < b]$$

**Ejemplo.**

Sea $f(x)$ una función definida por $$f(x) = \begin{cases}cx^2, & 0\leq x \leq 2; \\0, & \text{en otro caso}. \end{cases}$$

1. Calcule el valor de la constante $c$ para que $f$ sea la función de densidad para alguna variable aleatoria $X$.
2. Calcule $P[0 < X\leq 1]$.

**Solución.**

1. $\displaystyle 1 = \int_{-\infty}^{\infty} f(x)\,\mathrm{d}x = \int_{0}^{2}cx^2\,\mathrm{d}x \Longrightarrow c = 3/8$
2. $\displaystyle P[0 < X \leq 1] = \int_{0}^{1}\frac{3x^2}{8}\,\mathrm{d}x = \frac{1}{8}$

### 2.2. Valor esperado

Supongamos que $X$ es una variable continua. El **valor esperado de $X$** viene dado por

$$E[X] = \int_{-\infty}^{\infty} x f(x)\,\mathrm{d}x.$$

Se usará también $\mu$ para denotar al valor esperado de $X$, es decir $\mu = E[X]$.

**Propiedades**
- Sean $a,b$ dos constantes reales. Entonces $E[aX+b] = aE[X]+b$
- Sean $X$ e $Y$ dos variables aleatorias y $a,b$ dos constantes reales. Entonces $E[aX + bY] = aE[X] + bE[Y]$

### 2.3. Varianza

Supongamos que $X$ es una variable continua. La **varianza de $X$** viene dada

$$\begin{align*} \mathrm{Var}(X) &= \int_{-\infty}^{\infty} (x-E[X])^2 f(x)\,\mathrm{d}x \\ &= E[X^2] - \mu^2 \end{align*}$$

Se usará también $\sigma^2$ para denotar la varianza de $X$, es decir, $\sigma^2 = \mathrm{Var}(X)$. Al valor $\sigma = \sqrt{\mathrm{Var}(X)}$ se le denominará **desviación estándar de $X$**.

**Propiedades.**
- Sean $a,b$ son dos constantes reales. Entonces $\mathrm{Var}(aX+b) = a^2\mathrm{Var}(X)$.

**Ejemplo.**

El costo $C$ de un proyecto (en miles de dólares) está dado por $C = 3X + 5X^2$, donde $X$ es el tiempo empleado. La variable $X$ verifica

$$E[(X-0.5)^2] = 13/4, \quad E[(X-1)^2]=2.$$

1. Determine la media y la varianza de $X$.
2. Calcule el costo esperado del proyecto.

**Solución.**

1. De los datos, obtenemos $E[X^2] - E[X] = 3$ y $E[X^2]-2E[X]=1$ (2 ecuaciones lineales con 2 incógnitas). Luego $E[X]=2$ y $E[X^2] = 5$. Finalmente, $\mathrm{Var}(X) = E[X^2] - \mu^2 = 5 - 4 = 1$.
2. $E[C] = E[3X+5X^2] = 3E[X] + 5E[X^2] = 31$. Así, el costo esperado del proyecto es de 31,000 dólares.

## 3. Distribuciones continuas importantes

### 3.1. Distribución normal

Se dice que una variable aleatoria continua $X$ se **distribuye normalmente** si su función de densidad viene dada por

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$

En adelante, escribiremos $X \sim \mathrm{N}(\mu, \sigma^2)$.

Se tiene que

$$E[X] = \mu, \quad \mathrm{Var}(X) = \sigma^2$$

## Probability density function - PDF ##

Es el valor de:

$$f(x) = \frac{1}{\sigma\sqrt{2\pi}}e^{-\frac{1}{2}\left(\frac{x-\mu}{\sigma}\right)^2}$$



Captura de pantalla (1593).png

```python
import scipy.stats
scipy.stats.norm(0,1).pdf(-1) #pdf = probability density function
```

```python
import math
def normpdf(x, mean, sd):
    var = float(sd)**2
    denom = (2*math.pi*var)**.5
    num = math.exp(-(float(x)-float(mean))**2/(2*var))
    return num/denom

normpdf(-1,0,1)
```

### Propiedad reproductiva de la normal

Sean $X_1, X_2,\ldots, X_n$ variables aleatorias independientes tales que $X_i \sim \mathrm{N}(\mu_i,\sigma_i^2)$ para $i=1,2,\ldots,n$. La variable aleatoria

$$\phantom{\int} c_1X_1 + c_2X_2 + \ldots + c_nX_n \sim \mathrm{N}(c_1\mu_1 + c_2\mu_2+ \ldots + c_n\mu_n, c_1^2\sigma_1^2 + c_2^2\sigma_2^2+ \ldots + c_n^2\sigma_n^2),$$

donde $c_i \in \mathbb{R}$ para $i=1,2,\ldots,n$.

**Ejemplo.**

### 3.2. Distribución t

Una variable aleatoria continua $X$ tiene una distribución **t-student con $r$ grados de libertad**, en adelante $X \sim t(r)$, si su función de densidad viene dada por

$$f(x) = \frac{\Gamma((r+1)/2)}{\Gamma(r/2)\sqrt{\pi r}}\left( 1+\frac{x^2}{r}\right)^{-(r+1)/2}$$
